```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    ConfusionMatrixDisplay
)
```

```python
# ==========================
# DATOS PERSONALIZADOS
# ==========================

np.random.seed(924)

n = 3924

edad = np.random.randint(18, 81, n)

imc = np.random.normal(27, 5, n)

presion = np.random.randint(90, 191, n)

glucosa = np.random.randint(70, 221, n)

actividad = np.random.choice(
    ['Alta','Media','Baja'],
    n,
    p=[0.30,0.40,0.30]
)

df = pd.DataFrame({
    'Edad': edad,
    'IMC': imc,
    'Presion': presion,
    'Glucosa': glucosa,
    'Actividad': actividad
})

df.head()
```

```python
df["IMC_Alto"] = (df["IMC"] > 30).astype(int)

df["Hipertension"] = (df["Presion"] > 140).astype(int)

df["Riesgo_Metabolico"] = (
    (df["Glucosa"] > 126) &
    (df["IMC"] > 30)
).astype(int)

# Variable objetivo

df["Riesgo_Enfermedad"] = (
    (
        df["Hipertension"] +
        df["Riesgo_Metabolico"] +
        (df["Edad"] > 60).astype(int)
    ) >= 2
).astype(int)

df.head()
```

```python
# Eliminar duplicados

df = df.drop_duplicates()

# Codificar actividad

encoder = LabelEncoder()

df["Actividad"] = encoder.fit_transform(df["Actividad"])

# Escalar variables numéricas

scaler = MinMaxScaler()

df[["Edad","IMC","Presion","Glucosa"]] = scaler.fit_transform(
    df[["Edad","IMC","Presion","Glucosa"]]
)

df.head()
```

```python
df.to_csv("dataset_personal.csv", index=False)

print("Dataset guardado correctamente.")

print("Cantidad de registros:", len(df))
```

```python
X = df.drop("Riesgo_Enfermedad", axis=1)

y = df["Riesgo_Enfermedad"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=924
)

print("Entrenamiento:", X_train.shape)

print("Prueba:", X_test.shape)
```

```python
rf = RandomForestClassifier(random_state=924)

rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)
```

```python
lr = LogisticRegression(max_iter=1000)

lr.fit(X_train, y_train)

pred_lr = lr.predict(X_test)
```

```python
print("===== RANDOM FOREST =====")

print("Accuracy:", accuracy_score(y_test,pred_rf))

print("Precision:", precision_score(y_test,pred_rf))

print("Recall:", recall_score(y_test,pred_rf))

print("F1:", f1_score(y_test,pred_rf))

print()

print("===== REGRESIÓN LOGÍSTICA =====")

print("Accuracy:", accuracy_score(y_test,pred_lr))

print("Precision:", precision_score(y_test,pred_lr))

print("Recall:", recall_score(y_test,pred_lr))

print("F1:", f1_score(y_test,pred_lr))
```

```python
cm = confusion_matrix(y_test, pred_rf)

disp = ConfusionMatrixDisplay(cm)

disp.plot()

plt.title("Random Forest")

plt.show()
```

```python
cm = confusion_matrix(y_test, pred_lr)

disp = ConfusionMatrixDisplay(cm)

disp.plot()

plt.title("Regresión Logística")

plt.show()
```
